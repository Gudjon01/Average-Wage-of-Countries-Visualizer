```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import panel as pn
from panel.interact import interact
pn.extension()
```

```python
df = pd.read_csv("avg_wage.csv")
df = df.set_index("Country")
```

```python
country = df.index.values.tolist()
year = df.columns.tolist()
```

```python
def create_plot(country):
    y = df.loc[country].tolist()
    plot = plt.figure()
    plt.title("Average Wage of "+ country)
    plt.bar(year,y)
    plt.close(plot)
    return plot

selected = pn.widgets.Select(name = "Please Select a Country", options = country)
interact(create_plot,country = selected)
```
